```python
import numpy as np
import tkinter as tk
import random as rd
```

```python
colors = ["green", "green1", "green2","green3", "green4"]
```

```python
class Branch:
    def __init__(self, x, y, color, angle, length):
        self.x = x
        self.y = y
        self.color = color
        self.texture = cs.create_oval(x-2, y-2, x+2, y+2, fill=color)
        self.length = length
        self.angle = angle
        self.healh = 2 + rd.randint(-2, 2)
```

```python
def move_particle(branch):
    x_old = branch.x
    y_old = branch.y

    dx = branch.length * np.cos(branch.angle)
    dy = branch.length * np.sin(branch.angle)
    
    branch.x += dx
    branch.y += dy

    cs.create_line(x_old, y_old, x_old + dx, y_old + dy, fill=branch.color, width=2)
    cs.move(branch.texture, dx, dy)
```

```python
root= tk.Tk()
root.title("Двумерное дерево")

cs = tk.Canvas(root, width=800, height=800)
cs.pack()

particles = []
particles.append(Branch(500 / 2, 500 / 2, "blue", 0, 10))
steps = 0

del_angle = np.pi / 20

def motion():
    global steps
    buff = []
    for cell in particles:
        move_particle(cell)
        if cell.length <= 0 or cell.healh < 0: 
            
            buff.append(Branch(cell.x, cell.y, rd.choice(colors), cell.angle + del_angle + np.pi / 40, cell.length - 1))
            buff.append(Branch(cell.x, cell.y, rd.choice(colors), cell.angle + (del_angle + np.pi / 40) * 2, cell.length - 1))
            particles.remove(cell)
        cell.healh -= 1
        
    particles.extend(buff)
    steps += 1
    
    if steps < 30: root.after(200, motion)
    

motion()
root.mainloop()
```
