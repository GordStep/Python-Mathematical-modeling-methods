```python
import tkinter as tk
import numpy as np
```

```python
class Particle:
    def __init__(self, canvas, pos, vel, radius, mass, color):
        self.canvas = canvas
        self.pos = np.array(pos, dtype=float)
        self.vel = np.array(vel, dtype=float)
        self.radius = radius
        self.mass = mass
        self.id = canvas.create_oval(self.pos[0] - self.radius, self.pos[1] - self.radius, self.pos[0] + radius, self.pos[1] + radius, fill=color) 

    def move(self):
        self.pos += self.vel * dt
        self.canvas.move(self.id, self.vel[0] * dt, self.vel[1] * dt)

    def check_collision(self, bounds):
        xmin, xmax, ymin, ymax = bounds

        if self.pos[0] - self.radius <= xmin:
            self.pos[0] = xmin + self.radius
            self.vel[0] *= -1
        if self.pos[0] + self.radius >= xmax:
            self.pos[0] = xmax - self.radius
            self.vel[0] *= -1  
        if self.pos[1] - self.radius <= ymin:
            self.pos[1] = ymin + self.radius
            self.vel[1] *= -1
        if self.pos[1] + self.radius >= ymax:
            self.pos[1] = ymax - self.radius
            self.vel[1] *= -1

    def check_particle_collision(self, p_other):
        delta_pos = self.pos - p_other.pos
        dist = np.linalg.norm(delta_pos)

        if dist <= (self.radius + p_other.radius):
            delta_vel = self.vel - p_other.vel

            normal = delta_pos / np.linalg.norm(delta_pos)

            if np.dot(delta_vel, normal) > 0:
                return
            v_rel = np.dot(delta_vel, delta_pos)
            self.vel = self.vel = v_rel/dist**2 * delta_pos
            p_other.vel = p_other.vel + v_rel / dist**2 * delta_pos
```

```python
root = tk.Tk()
cs_width = 300
cs_height = 300

cs = tk.Canvas(root, width=cs_width, height=cs_height)
cs.pack()

dt = 0.01
bound1 = [0, cs_width, 0, cs_height]
p1 = Particle(cs, [50, 100], [150, -150], 20, 1, 'blue')
p2 = Particle(cs, [100, 50], [-100, 150], 20, 1, 'green')

def motion():
    p1.move()
    p2.move()
    
    p1.check_particle_collision(p2)
    
    p1.check_collision(bound1)
    p2.check_collision(bound1)

    
    
    root.after(5, motion)

motion()

root.mainloop()
```
