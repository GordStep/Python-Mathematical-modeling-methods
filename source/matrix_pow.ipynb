3. вводится матрица A размером $n \times n, 0<p<10$ натуральное
найти $B=A^p$

$A^n = \begin{pmatrix} a_{11} & ... & a_{1m} \\ ... & & \\ a_{n1} & ... & a_{nm} \end{pmatrix}^n = A \cdot A \cdot ... \cdot A$

```python
def MakeSquareMatrix(n, value=0):
    res = []
    for i in range(n):
        res.append([])
        for j in range(n):
            res[i].append(value)
    return res
```

```python
def MatrixMultiplication(A, B):
    result = MakeSquareMatrix(len(A)) 
    for i in range(len(A)):
        for j in range(len(A)):
            for k in range(len(B)):
                result[i][j] += A[i][k] * B[k][j]
    return result
```

```python
def MatrixPow(a, n):
    if n == 0:
        return MakeSquareMatrix(len(a), 1)
    R = a.copy()
    
    for i in range(n - 1):
        R = MatrixMultiplication(R, a)
    return R
```

```python
A = [[4, 2, 1],
     [4, 0, 4],
     [0, 4, 1]]
B = [[4, 8, 1],
     [0, 5, 6],
     [2, 0, 5]]
D = A.copy()

MatrixMultiplication(A, B), MatrixPow(A, 1), MatrixPow(A, 2)
```
