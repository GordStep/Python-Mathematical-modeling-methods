Найдите одно из решений системы уравнений методом Ньютона с абсолютной погрешностью 0,0001.
$$ \begin{cases} x^3 + y^3 = 8 \\
                 y = 1 + x^{3/2}
    \end{cases} $$


```python
import math
```

```python
def f_1(vec_x):
    return vec_x[0]**3 + vec_x[1]**3 - 8
def f_2(vec_x):
    return 1 + vec_x[0]**1.5 - vec_x[1]
def F(vec_x):
    return (f_1(vec_x), f_2(vec_x))

def dF_x(vec_x, h=0.01):
    x = vec_x[0]
    y = vec_x[1]
    df_1 = (f_1((x + h, y)) - f_1(vec_x)) / h
    df_2 = (f_2((x + h, y)) - f_2(vec_x)) / h
    return (df_1, df_2)

def dF_y(vec_x, h=0.01):
    x = vec_x[0]
    y = vec_x[1]
    df_1 = (f_1((x, y + h)) - f_1(vec_x)) / h
    df_2 = (f_2((x, y + h)) - f_2(vec_x)) / h
    return (df_1, df_2)
```

```python
def moduleF(vec):
    x, y = F(vec)
    return math.sqrt(x**2 + y**2)
```

```python
def MethodNewton(vec_0, epsilon=0.0001):
    vec = vec_0
    while abs(F(vec)[0]) > epsilon and abs(F(vec)[1]) > epsilon:
        
        a_11 = dF_y(vec)[1]
        a_12 = -dF_y(vec)[0]
        a_21 = -dF_x(vec)[1]
        a_22 = dF_x(vec)[0]
        
        d = a_22 * (-a_11) - (-a_12) * a_21
        
        J = [[a_11 / d, a_12 / d],
             [a_21 / d, a_22 / d]]
        
        vec_x = vec[0] - 1 / d * (a_11 * F(vec)[0] + a_12 * F(vec)[1]) 
        vec_y = vec[1] - 1 / d * (a_21 * F(vec)[0] + a_22 * F(vec)[1]) 
        
        vec = (vec_x, vec_y)
    return vec
        
```
