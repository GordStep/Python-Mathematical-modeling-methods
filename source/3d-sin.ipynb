```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
x = np.linspace(-20, 20, 1000)
y = x
x, y = np.meshgrid(x,y)
z = np.sin(np.sqrt(x**2 + y**2) / 2) + np.sin(np.sqrt(x**2 + y**2) / 4) + np.sin(np.sqrt(x**2 + y**2))
```

```python
fig = plt.figure()
f = fig.add_subplot(111, projection='3d')
f.plot_surface(x, y, z, label='sin')
f.set(zlim=(0, 10))

plt.show()
fig.savefig("pic.png")
```
