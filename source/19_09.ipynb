```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#dT/dt = -r(T - Ts)
```

```python
t_0 = 0
t_max = 60
dt = 0.1
N = int(t_max / dt)
t = np.arange(t_0, t_max, dt)

Ts = 22 # t окр среды
T_0 = 83
r = 0.1

T = np.zeros(N)
T[0] = 83
```

```python
T_a = Ts + np.exp(-r * t) * (T_0 - Ts)

plt.plot(t, T_a, 'g')

# метод Эйлера
for i in range(N - 1):
    T[i + 1] = T[i] - r*(T[i] - Ts) * dt

plt.plot(t, T, 'b--')

# Метод Рунге-Кутта
T_r = np.zeros(N)
T_r[0] = 83

for i in range(N - 1):
    k_1 =  -r * (T_r[i] - Ts) * dt
    k_2 =  -r * (T_r[i] + k_1 / 2 - Ts) * dt
    k_3 =  -r * (T_r[i] + k_2 / 2 - Ts) * dt
    k_4 =  -r * (T_r[i] + k_3 - Ts) * dt

    T_r[i + 1] = T_r[i] + (k_1 + 2 * k_2 + 2 * k_3 + k_4) / 6
    
plt.plot(t, T_r, 'r')

plt.legend(['Эталон', 'Эйлер', 'Рунге-Кутт'])
```

```python
Ads_err_euler = np.max(np.abs(T_a - T))
print(Ads_err_euler)

Ads_err_rang = np.max(np.abs(T_a - T_r))
print(Ads_err_rang)
```
