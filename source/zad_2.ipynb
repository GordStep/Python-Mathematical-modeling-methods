Выполните отделение корней графически и найдите один из корней методами простой итерации и Ньютона
с абсолютной погрешностью 0,0001.
$$7^x - 6x = 2$$

Метод дихотомии:
$$f(x) = 0, \qquad f(x) = 7^x - 6x - 2 = 0$$
$$ x + f(x) = 7^x - 5x -2 = \Phi(x) = x $$

$$x^0 = 1, x^1 = \Phi(x^0), x^2 = \Phi(x^1), ..., x^n = \Phi(x^{n-1})$$

Остановка по принципу:
$$|x^n - x^{n-1}| < \Epsilon \quad или \quad |f(x^n) < \Epsilon $$

```python
import math
```

```python
def Phi(x):
    return 7**x - 5 * x - 2

def f(x):
    return 7**x - 6 * x - 2
```

```python
def MethodSimpleIter(x_0, epsilon=0.0001):
    x = x_0
    while (abs(f(x)) > epsilon):
        x = Phi(x)
        print(x, f(x), Phi(x))
    return x
```

```python
MethodSimpleIter(-0.08)
```

```python
def dPhi(x, h=0.001):
    return (Phi(x + h) - Phi(x)) / h
```

```python
count_plus = 0
for x in range(-200, 200):
    if dPhi(x / 400) > 0:
        count_plus += 1
        print(x/400, "+")
    else:
        print(x/400, "-")
count_plus    
```

```python
def dF(x, h=0.001):
    return (f(x + h) - f(x)) / h
```

```python
def MethodNewton(x_0, epsilon=0.001):
    x = x_0
    while abs(f(x)) > epsilon:
        x = x - f(x) / dF(x)
        print(x)
    return x
```
