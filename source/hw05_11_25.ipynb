## (Д/з) Метод Проб и ошибок:
$$x^2 + y^2 + z^2 = R^2 = 1$$
Строим $n$ точек в объёме, проверяем попадает ли точка в объём и по формуле

$$S = S_0 \times \frac MN $$
$M$ - кол-во точек внутри фигуры
$N$ - кол-во точек всего
$S_0$ - площадь между прямыми

```python
import numpy as np
```

```python
# Границы объёма сферы
R = 1
a = -R
b = R

V_ab = (b - a)**3

V_th = 4 / 3 * np.pi * R**3 
n = 1000
```

```python
def isIn(x_, y_, z_):
    return x_**2 + y_**2 + z_**2 <= R
```

```python
x_rand = np.random.uniform(a, b, n)
y_rand = np.random.uniform(a, b, n)
z_rand = np.random.uniform(a, b, n)
```

```python
M = sum([isIn(x, y, z) for x, y, z in zip(x_rand, y_rand, z_rand)])
M
```

```python
V_pr = V_ab * M / N
```
