```python
import numpy as np
import tkinter as tk
import random as rd
```

```python
class Particle:
    def __init__(self, x, y, color, delta_angle):
        self.x = x
        self.y = y
        self.color = color
        self.texture = cs.create_oval(x-2, y-2, x+2, y+2, fill=color)
        self.healh = 10 + rd.randint(-4, 5)
        self.delta_angle = delta_angle
```

```python
def move_particle(particle):
    particle.healh -= 1
    move_dist = 10
    move_angle = rd.uniform(0, np.pi) + particle.delta_angle

    dx = move_dist * np.cos(move_angle)
    dy = move_dist * np.sin(move_angle)

    if particle.x > 500:
        particle.x = 0
    if particle.x < 0:
        particle.x = 500
    if particle.y + dy > 500:
        particle.y = 0
    if particle.y < 0:
        particle.y = 500
        
    x_old = particle.x
    y_old = particle.y
    
    particle.x += dx
    particle.y += dy

    cs.create_line(x_old, y_old, x_old + dx, y_old + dy, fill=particle.color, width=2)
    cs.move(particle.texture, dx, dy)
```

```python
colors = ["red", "orange", "black", "green", "blue", "indigo", "violet", "maroon"]
```

```python
root= tk.Tk()
root.title("Двумерное случайное блуждание")

cs = tk.Canvas(root, width=500, height=500)
cs.pack()

particles = []
particles.append(Particle(500 / 2, 500 / 2, "blue", 0))
              
def motion():
    for cell in particles:
        move_particle(cell)

        if cell.healh == 0:
            ivent = rd.randint(0, 5)
            
            if ivent == 1 or len(particles) == 1:
                particles.append(Particle(cell.x, cell.y, cell.color, cell.delta_angle - np.pi / 2))
                particles.append(Particle(cell.x, cell.y, cell.color, cell.delta_angle + np.pi / 2))
            elif ivent > 1:
                particles.append(Particle(cell.x, cell.y, rd.choice(colors), cell.delta_angle))        
            
            particles.remove(cell)

    
    if len(particles) != 0:
        root.after(200, motion)
    

motion()
root.mainloop()
```
