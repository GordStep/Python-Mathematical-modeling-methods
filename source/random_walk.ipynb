```python
import numpy as np
import tkinter as tk
import random as rd
```

```python
root= tk.Tk()
root.title("Двумерное случайное блуждание")

cs = tk.Canvas(root, width=500, height=500)
cs.pack()

x, y = 500 / 2, 500 / 2
particle = cs.create_oval(x-2, y-2, x+2, y+2, fill="blue")

x_line = x
y_line = y

def motion():
    global x, y, x_line, y_line
    dx, dy = 0, 0
    x_line = x
    y_line = y
    r = rd.uniform(1, 5)
    phi = rd.uniform(0, 2 * np.pi)
    
    
    dx = r * np.cos(phi)
    dy = r * np.sin(phi)

    x += dx
    y += dy
    
    
    cs.create_line(x_line, y_line, x_line + dx, y_line + dy, fill="silver", width=1)
    cs.move(particle, dx, dy)
    root.after(20, motion)
    

motion()
root.mainloop()
```
