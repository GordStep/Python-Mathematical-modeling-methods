```python
import numpy as np
import tkinter as tk
```

```python
w = 600
h = 600

x0 = w / 2 + 100
y0 = h / 2 

maxIteration = 1000
Rad = 2
```

```python
def rgb_color(color):
    return '#%02x%02x%02x' % color
```

```python
root = tk.Tk()
root.title('Фрактал Мандельброта')
cs = tk.Canvas(root, width=w, height=h, bg='white')
cs.pack()

for x in range(w):                                                                                                                                                                                                                
    for y in range(h):
        a = (x - x0) / (w / 4)
        b = (y0 - y) / (w / 4)
        
        c = complex(a, b)
        z = complex(0)
        n = 0
        for n in range(maxIteration):
            z = z**2 + c
            if abs(z) > 2:
                break

        if n == maxIteration - 1:
            col = rgb_color((0, 0, 0))
        else:
            col = rgb_color((255, 255, 255))
        cs.create_oval(x, y, x+1, y+1, outline=col)

root.mainloop()
```
