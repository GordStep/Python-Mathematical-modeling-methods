```python
import numpy as np
import tkinter as tk
```

```python
w = 600
h = 600

x0 = w / 2
y0 = h / 2

p = -0.8
q = 0.156
maxRez = 1.5
maxIteration = 100
Rad = 2
```

```python
def rgb_color(color):
    return '#%02x%02x%02x' % color
```

```python
root = tk.Tk()
cs = tk.Canvas(root, width=w, height=h, bg='white')
cs.pack()

for x in range(w):
    for y in range(h):
        Rez = maxRez / x0 * (x - x0)
        Imz = maxRez / y0 * (y - y0)

        for i in range(maxIteration):
            Rez_old = Rez
            Imz_old = Imz

            Rez = Rez_old**2 - Imz_old**2 + p + x0
            Imz = 2 * Rez_old * Imz_old + q + y0

            if Rez**2 + Imz**2 > Rad**2:
                col = rgb_color((0, i * 9 % 255, i * 9 % 255))
                break
        cs.create_oval(x, y, x, y, outline=col)

root.mainloop()
```
