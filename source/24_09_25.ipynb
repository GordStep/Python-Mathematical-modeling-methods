```python
import numpy as np
import matplotlib.pyplot as plt
```

$a_x = - \frac{GMx}{r^3}$

$V_x = a_x t$

$V_y = a_y t$

$x = V_x t$

$y = V_y t$

$\vec{F_{т}} = - \frac{GMm}{r^3} \vec{r}$

${F_{т}}_x = - \frac{GMm}{r^2} \cos \theta = m a_x$

$\cos \theta = \frac xr$

$\sin \theta = \frac yr$

$ V_x = V \sin \theta = \frac{Vy}r$

$ V_y = V \cos \theta = \frac{Vx}r$

$$ V = \sqrt{GM \left(\frac2r - \frac1a\right)} $$

```python
GM = 4 * np.pi**2 # в усл единицах
dt = 0.01
N = 200

# Нач усл
x = 1
y = -1
r = np.sqrt(x**2 + y**2)
v = np.sqrt(GM / r)
vx = np.abs(v * y / r)
vy = np.abs(v * x / r)
```

```python
for i in range(N):
    ax = -GM * x / (r**3)
    ay = -GM * y / (r**3)
    vx += ax * dt
    vy += ay * dt

    x += vx * dt
    y += vy * dt
    
    plt.plot(x, y, 'bo')
plt.plot(0,0,'yo', markersize=10)
plt.axis('equal')
```
