# Метод Гаусса


```python
import numpy as np
%matplotlib inline
```

1. F-норма
$$||A||_F = \sqrt{ \sum_{i = 1}^m \sum_{j = 1}^n |A_{ij}|^2 } $$

2. Норма $L_1$
$$ ||A||_1 = \max \sum_{i = 1}^m |A_{ij}|^2, \quad j \in \overline{1, n} $$

2. Норма $L_\infty$
$$ ||A||_\infty = \max \sum_{j = 1}^n |A_{ij}|^2, \quad i \in \overline{1, m} $$

```python
def F_norm(A):
    f_norm = 0
    
    for i in range(len(A)):
        for j in range(len(A[0])):
            f_norm += A[i][j]**2
    return f_norm
```

```python
n = 19
slau_matrix = np.zeros((n+1, n+1))

slau_matrix[0][0] = 1
slau_matrix[n][0] = 1
slau_matrix[n][n] = 1

for i in range(1, n):
    slau_matrix[i][i] = -2
    slau_matrix[i][i - 1] = 1
    slau_matrix[i][i + 1] = 1
for i in range(1, n):
    slau_matrix[n][i] = 2

slau_matrix
```

```python
inv = np.linalg.inv(slau_matrix)
np.linalg.norm(slau_matrix), np.linalg.norm(inv), np.linalg.eig(slau_matrix)
```

```python
f_matrix = np.zeros((n+1, 1))

f_matrix[0] = 1
f_matrix[n] = 1

for i in range(1, n):
    f_matrix[i] = 2 / ((i + 1)**2)

f_matrix
```

# QR-алгортим

## Классический процесс Грама — Шмидта
**Алгоритм**

Пусть имеются линейно независимые векторы $a_1, \dots , a_n$ и пусть \
$\text{proj}_b a$ — оператор проекции вектора $a$ на вектор $b$, определённый как
$$
\text{proj}_b a = \frac{(a , b)}{(b, b)} b
$$

где $(a, b)$ — скалярное произведение векторов $a$ и $b$.

Классический процесс Грама — Шмидта выполняется следующим образом:
$$
b_1 = a_1 \\
b_2 = a2 - \text{proj}_{b_1} a_2 \\
b_2 = a3 - \text{proj}_{b_1} a_3 - \text{proj}_{b_2} a_3 \\
\dots \\
b_n = a_n - \text{proj}_{b_1} a_n - \text{proj}_{b_2} a_n - \cdots - \text{proj}_{b_{n-1}} a_n
$$

На основе каждого вектора $b_j, j = 1...n$ может быть получен нормированный вектор $e_j$ единичной длины, определённый как
$$
e_j = \frac{b_j}{||b_j||} = \frac{b_j}{\sqrt{(b_j, b_j)}}
$$

Результаты процесса Грама — Шмидта:

$e_1 ,\dots , e_n$ - система ортонормированных векторов

```python
def scalarMult(a, b):
    res = 0
    for i in range(len(a)):
        res += a[i] * b[i]
    return res

def proj(b, a):
    coef = scalarMult(a, b) / scalarMult(b, b)

    return  [coef * el for el in a]
```

```python
def algGramSchmidt(a):
    vectors = [np.array(v, dtype=float) for v in a]
    orthogonal = []
    
    for v in vectors:
        u = v.copy()
        
        for w in orthogonal:
            # Проекция v на w
            proj = np.dot(v, w) / np.dot(w, w) * w
            u = u - proj
        
        if np.linalg.norm(u) > 1e-10:
            u = u / np.linalg.norm(u)
            orthogonal.append(u)
    
    return orthogonal
```

# QR-Разложение
Матрица $A$ размера $n \times m$, где $n \geq m$, с комплексными элементами может быть представлена в виде

   $$ A = QR $$
    

```python
def qr_decomposition(A):
    vectors_column = []

    for i in range(len(A)):
        vectors_column.append([0]*len(A))
    print(len(vectors_column), len(vectors_column[0]))
    for i in range(len(A)):
        for j in range(len(A)):
            vectors_column[i][j] = A[j][i]
    
    e_vects = algGramSchmidt(vectors_column)

    Q = np.zeros((len(A), len(A)))

    for i in range(len(A)):
        for j in range(len(A)):
            Q[i][j] = e_vects[j][i]
    
    R = np.dot(np.linalg.matrix_transpose(Q), A)

    return Q, R
    
```

```python
def qr_algorithm(A, num_iter, eps):
    T = A.copy()
    U = np.eye(A.shape[0])
    conv = [(T, U)]
    for i in range(num_iter):
        Q, R = qr_decomposition(T)
        T = R @ Q
        U = U @ Q
        conv.append((T, U))
        if np.sum(np.abs(np.tril(T, k=-1))) < eps:
            break
    return T, U, conv[1:]
```

```python
def getDiscrepancyVector(A, X, F):
    return np.dot(A, X) - F
```

```python
def MethodGaussa(A_matrix, F_matrix):
    A = A_matrix.copy()
    F = F_matrix.copy()
    
    for i in range(len(A)):
        for j in range(i + 1, len(A)):
            coef = A[j][i] / A[i][i]
            for k in range(len(A)):
                A[j][k] -= A[i][k] * coef
            F[j] -= F[i] * coef
                
    X = np.zeros((len(A), 1))
    
    for i in range(len(A) - 1, -1, -1):
        x_sum = F[i]
        for j in range(i, len(A)):
            x_sum -= A[i][j] * X[j]
        X[i] = x_sum / A[i][i]
    
    r = getDiscrepancyVector(A_matrix, X, F_matrix)
    print(r)
    
    return X
```

```python
def MethodJacobiGauss(A_matrix, F_matrix, Eps=0.001):
    A = A_matrix.copy()
    F = F_matrix.copy()
    n = len(A)

    L = np.zeros((n, n))
    D = np.zeros((n, n))
    U = np.zeros((n, n))

    for i in range(n):
        for j in range(i):
            L[i][j] = A[i][j]
    for i in range(n):
        D[i][i] = A[i][i]
        
    for i in range(n):
        for j in range(i + 1, n):
            U[i][j] = A[i][j]

    R = np.dot(-1 * np.linalg.inv(D), L + U)
    F_vect = np.dot(np.linalg.inv(D), F)

    X = np.zeros((n, 1))
    X_next = np.dot(R, X) + F_vect
    
    r = getDiscrepancyVector(A, X_next, F) # вектор невязки
    
    # Eps = 0.001
    while max(r) > Eps:
        X = X_next
        X_next = np.dot(R, X) + F_vect
        
        r = getDiscrepancyVector(A, X_next, F)
        # print(r)
    r = getDiscrepancyVector(A, X_next, F)
    print(r)
    return X
    
```
