```python
import numpy as np
import tkinter as tk
```

```python
is_paused = False
```

```python
def toggle_pause():
    global is_paused
    is_paused = not is_paused

    # root.grab_set()

def close_window():
    root.destroy()
```

```python
dots = []
```

```python
GM = 4 * np.pi**2 # в усл единицах
dt = 0.01
N = 200

g = 9.81 # м/с
M = 2 # кг

WIDTH = 1200
HEIGHT = 700

# Нач усл
x = 0
y = 0
r = 2

vx_0 = 25
vy_0 = -30

vx = vx_0
vy = vy_0
```

```python
root = tk.Tk()
cs = tk.Canvas(root, width=WIDTH, height=HEIGHT)
cs.pack()
```

```python
# # Рисуем центральный шар
# x_scr0 = 500 / 2
# y_scr0 = 500 / 2
# rads = 10
# cs.create_oval(x_scr0 - rads, y_scr0 - rads, x_scr0 + rads, y_scr0 + rads, fill='yellow')
```

```python
# Рисуем движующийся шар
radp = 5
coef = 10
x_scr = x * coef
y_scr = HEIGHT + y * coef
ball = cs.create_oval(x_scr - radp, y_scr - radp, x_scr + radp, y_scr + radp, fill='gray')
```

```python
count_frames  = 0

time = 0
```

```python
def proizvod_y(dot1, dot2):
    dy = 0

    
    dy = ((dot2[1] - dot1[1]) / (dot2[0]- dot1[0]))

    return dy
```

```python
x_line = x_scr
y_line = y_scr

del_y_old = 0


def motion():
    if is_paused:
        return
    
    global x, y, vx, vy, r, x_line, y_line, count_frames, time, del_y_old
    ax = 0
    ay = M * g
    vx += ax * dt
    vy += ay * dt

    dx = vx * dt
    dy = vy * dt
    x += dx
    y += dy

    dots.append((x, y))
    count_frames += 1

    if len(dots) > 2:
        del_y = proizvod_y(dots[-2], dots[-1])
        cs.create_line(dt * count_frames, del_y_old, dt * count_frames + dt * coef, del_y_old + del_y * coef, fill="silver", width=1) 

        del_y_old = del_y
    
    cs.move(ball, dx*coef, dy*coef)
    cs.create_line(x_line, y_line, x_line + dx * coef, y_line + dy * coef, fill="silver", width=1) 

    

    root.after(20, motion)
    r = np.sqrt(x**2 + y**2)

    

    x_line = x_line + dx * coef
    y_line = y_line + dy * coef

    print(f'{x=}, {y=}')

    if y > 0:
        time = count_frames * dt

        toggle_pause()
        root.after(2000, close_window())
        # root.quit()


motion()
root.mainloop()
```

```python
def proiz(dots):
    dy = []

    for i in range(1, len(dots)):
        dy.append((dots[i][1] - dots[i - 1][1]) / (dots[i][0]- dots[i - 1][0]))

    return dy
```

```python
import matplotlib.pyplot as plt

t = np.linspace(0, time, count_frames - 1)

y_p = proiz(dots)

plt.plot([0, 0], [min(y_p), max(y_p)], color='gray', linestyle='--', linewidth=0.9, alpha=0.5)
plt.plot([0, time], [0, 0], color='gray', linestyle='--', linewidth=0.9, alpha=0.5)
plt.plot(t, y_p)
```
