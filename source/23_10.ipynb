# Метод Монте-Карло
Общее название круппы методов, основанных на большом случайном шумом

$$M(X) = A$$

$$x = a + (b - a) \times r$$
Где $[a, b]$ - диапазон, $r$ - случайное число $[0, 1]$ 

$M$ - кол-во точек внутри фигуры
$N$ - кол-во точек всего
$S_0$ - площадь между прямыми

$$S = S_0 \times \frac MN $$

## (Д/з) Метод Проб и ошибок:
$$x^2 + y^2 + z^2 = R^2 = 1$$
Строим $n$ точек в объёме, проверяем попадает ли точка в объём и по формуле

```python
import random as rand
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0
b = 1
n = 1_000

def F(x_):
    return 4 * np.sqrt(1 - x_**2)

x = np.linspace(0, 1, 100)
y = F(x)

plt.plot(x, y, color="red")

a1 = 0
b1 = 4

x_rand_mass = np.random.uniform(a, b, n)
y_rand_mass = np.random.uniform(a1, b1, n)

plt.scatter(x_rand_mass, y_rand_mass)
dots_in = 0
for i in range(n):
    if y_rand_mass[i] <= F(x_rand_mass[i]):
        dots_in += 1

S = (b - a) * (b1 - a1) * (dots_in / n)

print(S)
```

```python
x_rand = np.random.uniform(a, b, n)
y_rand = F(x_rand)

S_sr = (b - a) * sum(y_rand) / n
print(S_sr)
```
