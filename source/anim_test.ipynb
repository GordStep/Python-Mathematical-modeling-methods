```python
import numpy as np
import tkinter as tk
```

```python
GM = 4 * np.pi**2 # в усл единицах
dt = 0.01
N = 200

# Нач усл
x = 1
y = -1
r = np.sqrt(x**2 + y**2)
v = np.sqrt(GM / r) * 0.5
vx = np.abs(v * y / r)
vy = np.abs(v * x / r)
```

```python
root = tk.Tk()
cs = tk.Canvas(root, width=500, height=500)
cs.pack()
```

```python
# Рисуем центральный шар
x_scr0 = 500 / 2
y_scr0 = 500 / 2
rads = 10
cs.create_oval(x_scr0 - rads, y_scr0 - rads, x_scr0 + rads, y_scr0 + rads, fill='yellow')
```

```python
# Рисуем движующийся шар
radp = 5
coef = 100
x_scr = x_scr0 + x * coef
y_scr = y_scr0 + y * coef
ball = cs.create_oval(x_scr - radp, y_scr - radp, x_scr + radp, y_scr + radp, fill='gray')
```

```python
x_line = x_scr
y_line = y_scr
def motion():
    global x, y, vx, vy, r, x_line, y_line
    ax = -GM * x / (r**3)
    ay = -GM * y / (r**3)
    vx += ax * dt
    vy += ay * dt

    dx = vx * dt
    dy = vy * dt
    x += dx
    y += dy
    
    cs.move(ball, dx*coef, dy*coef)
    cs.create_line(x_line, y_line, x_line + dx * coef, y_line + dy * coef, fill="silver", width=1) 
    root.after(20, motion)
    r = np.sqrt(x**2 + y**2)

    x_line = x_line + dx * coef
    y_line = y_line + dy * coef

motion()
root.mainloop()
```
